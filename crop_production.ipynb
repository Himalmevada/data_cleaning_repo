```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("crop_production.csv")
```

```python
df.head()
```

### 1.get Overall Crop Production By state.and note down the conclusions.

```python
df.groupby("State_Name")["Production"].sum().sort_values(ascending=False)
```

```python
# Kerla , Aandhrapradesh and Tamilnadu is on the top for state wise production
```

### 2.visualize the productivity of  different states and note down the conclusions.

```python
top_five_state_prod_wise = list(df.groupby("State_Name")["Production"].sum().sort_values(ascending=False)[:5].index)
```

```python
for state in top_five_state_prod_wise:
    state_filter = df[df["State_Name"] == state]
    state_grp = state_filter.groupby("Season")["Production"].sum()
    state_grp.plot(kind="bar")
    plt.title(f"Season Wise Production of State : {state}")
    plt.show()
```

```python
# Most of the first highest production happen in whole year and then in second place it is Kharif season. 
```

### 3.get Overall production through years.note down the conclusions.

```python
df.groupby("Crop_Year")["Production"].sum().sort_values(ascending=False)
```
